```python
import geopandas as gpd
```

```python
points = gpd.GeoDataFrame({
    'ids': ['a', 'b', 'c'],
    'geometry': gpd.points_from_xy([.0001, .1, .00003], [.0004, .0004, .0004])})
points = points.set_crs(epsg=4326)
points = points.to_crs(epsg=3857)
```

```python
points
```

```python
points_buffered = points.copy()
points_buffered['geometry'] = points_buffered.geometry.buffer(1e4)
```

```python
joined = gpd.sjoin(points_buffered, points, how='inner', predicate='intersects')
joined.columns
```

```python
joined_drop = joined.loc[joined['ids_left'] != joined['ids_right']]
deduplicated_points = points.copy()
groups = joined_drop.groupby('ids_left')

ids_to_drop = joined_drop['ids_right'].unique().tolist()
for _, group in groups:
    ids_to_drop = [id for id in ids_to_drop if id != group['ids_left'].values[0]]
    deduplicated_points = deduplicated_points.loc[~deduplicated_points['ids'].isin(ids_to_drop)]


deduplicated_points
```
